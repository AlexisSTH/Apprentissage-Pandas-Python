```python
import pandas as pd

employes = pd.DataFrame({
    'nom': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'departement': ['IT', 'RH', 'IT', 'Finance', 'RH', 'IT'],
    'age': [28, 35, 42, 31, 26, 38],
    'salaire': [45000, 52000, 67000, 58000, 41000, 61000],
    'annees_exp': [3, 8, 15, 6, 2, 12]
}, index=['E001', 'E002', 'E003', 'E004', 'E005', 'E006'])

print(employes)
```

```python
# Exercice 11
# MÃ©thode 1 : Utiliser loc aprÃ¨s avoir sÃ©lectionnÃ© les lignes
employes.iloc[1:4].loc[:, ['age', 'salaire']]
```

```python
# Exercice 12 Augmenter salaire IT
employes.loc[employes['departement'] == 'IT', 'salaire'] += 5000
employes
```

```python
# Exercice 13- Nom du salaire max
employes.loc[employes['salaire'].idxmax(), 'nom']
```

```python
# Exercice 14 - ReorganisÃ© les colonnes
employes.loc[:, ['salaire', 'nom', 'departement']]
```

```python
# Exercice 15 - colonne niveau
employes.loc[employes['annees_exp'] >= 10, 'niveau'] = 'Senior'
employes.loc[employes['annees_exp'] < 10, 'niveau'] = 'Junior' 
employes
```

```python
# Exercice 16 - condition multiple (plus de 30 ans) gagne plus de 5000â‚¬ (Sont dans le dÃ©partement IT ou Finance)
employes.loc[
    (employes['age'] > 30) &
    (employes['salaire'] > 5000) &
    (employes['departement'].isin(['IT', 'Finance']))
]
```

```python
# Exercice 17 - Afficher RH et augmenter  le salaire

#afficher les noms
print(employes.loc[employes['departement'] == 'RH', 'nom'])

# Augmenter de 10%
employes.loc[employes['departement'] == 'RH', 'salaire'] *= 1.10
employes
```

```python
# Exercice 18 - une ligne sur deux, colonnes paire
employes.iloc[1::2, ::2]
```
