```python
# Exercice 11 Colonne rang_q3
ventes['rang_q3'] = ventes['ventes_q3'].rank(ascending=False)
ventes
```

```python
# Exercice 12 Top 3 ventes_q2
ventes.sort_values('ventes_q2', ascending=False).head(3)
```

```python
# Exercice 13 Rang moyen sur 3 trimestres
rang_q1 = ventes['ventes_q1'].rank(ascending=False)
rang_q2 = ventes['ventes_q2'].rank(ascending=False)
rang_q3 = ventes['ventes_q3'].rank(ascending=False)

rang_moyen = (rang_q1 + rang_q2 + rang_q3) / 3
print(rang_moyen)
```
